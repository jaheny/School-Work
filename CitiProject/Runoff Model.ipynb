```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
custtof_log= pd.read_csv('custtof_log.csv')
custtof_logd1= pd.read_csv('custtof_logd1.csv')
monthly_1m_logd3= pd.read_csv('monthly_1m_logd3.csv')
mon08= pd.read_csv('mon08.csv')
mon11= pd.read_csv('mon11.csv')
mon12= pd.read_csv('mon12.csv')
```

```python
custtof_log.head()
```

```python
decayRate = -0.0638938080181719 + 0.0128018391611731 * custtof_log + 0.142619991785448 * custtof_logd1 - 0.00186939153346944 * monthly_1m_logd3 - 0.00447840719738155 * mon08 - 0.0032027133009468 * mon11 - 0.00111520928765421 * mon12
```

```python
decayRate.head()
```

```python
inputBalance = pd.read_csv('inputBalanceOfRunoffModelMMDA.csv') #MMDA
```

```python
rowNumber = decayRate.shape[0] - 1
columnNumber = decayRate.shape[1]
forecastBalance = {}
for i in range(rowNumber):
    forecastBalance[i] = []
    for j in range(columnNumber):
        if j == 0:
            forecastBalance[i].append(inputBalance["balance"][i] * np.exp(decayRate.iloc[i][j]))
        else:
            forecastBalance[i].append(forecastBalance[i][-1] * np.exp(decayRate.iloc[i][j]))
```

```python
#TTS
monthly_1m_to_max1ml_dif2 = pd.read_csv('monthly_1m_to_max1ml_dif2.csv')
accttof01 = pd.read_csv('accttof01.csv')
accttof02 = pd.read_csv('accttof02.csv')
accttof03 = pd.read_csv('accttof03.csv')
mon02 = pd.read_csv('mon02.csv')
mon03 = pd.read_csv('mon03.csv')
```

```python
decayRate = - 0.00902596453067748 - 0.01922747596207270 * monthly_1m_to_max1ml_dif2 - 0.15181632687163300 * accttof01 \
- 0.03712068578206740 * accttof02 - 0.00704908918502350 * accttof03 - 0.00314978995891553 * mon02 + 0.00339523876260364 * mon03
```

```python
decayRate.head()
```

```python
inputBalance = pd.read_csv('inputBalanceOfRunoffModelTTS.csv')
```

```python
rowNumber = decayRate.shape[0] - 1
columnNumber = decayRate.shape[1]
forecastBalance = {}
for i in range(rowNumber):
    forecastBalance[i] = []
    for j in range(columnNumber):
        if j == 0:
            forecastBalance[i].append(inputBalance["balance"][i] * (1 + decayRate.iloc[i][j]))
        else:
            forecastBalance[i].append(forecastBalance[i][-1] * (1 + decayRate.iloc[i][j]))
```
